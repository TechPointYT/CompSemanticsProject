```python
import pandas as pd
from collections import Counter
import os


def convertData_pickle(fileName, csvCat, curFileIdx):
    Words = []
    lemma = []
    pos = []
    full_file_df = pd.DataFrame([])
    with open(fileName) as f:
        for line in f:
            info = line.split()
            Words.append(info[0])
            lemma.append(info[1])
            pos.append(info[2])
        f.close()
    lemma_counts= dict(Counter(lemma))
    pos_counts = dict(Counter(pos))
    lemma_counts_df = pd.DataFrame({"lemma":list(lemma_counts.keys()),"lemma_counts ": list(lemma_counts.values()) })
    lemma_counts_df.set_index("lemma", inplace=True)
    full_file_df = full_file_df.append(lemma_counts_df, sort=True)
    
    pos_counts_df = pd.DataFrame({"pos":list(pos_counts.keys()), "POS":list(pos_counts.values())})
    pos_counts_df.set_index("pos", inplace=True)
    full_file_df = full_file_df.append(pos_counts_df, sort=True)
    full_file_df = full_file_df.T
    full_file_df.to_csv(f".\\pickle_dfs\\{csvCat}_{curFileIdx}.csv")



rootdir = "C:\\Users\\David\\Desktop\\CompSemanticsProject\\text_files_centuries" 
centuries= ["18","19", "20"]
currentCentIdx= 0
curNumfile = 0
currentCent = centuries[0]

for subdir, dirs, files in os.walk(rootdir):
    curNumFile = 0 
    for file in files:
        print(f"{currentCent}{curNumFile}")
        if currentCent not in file:
            currentCentIdx += 1 
            currentCent = centuries[currentCentIdx]
        convertData_pickle(os.path.join(subdir, file), currentCent, curNumFile)
        curNumFile += 1    
    




```
