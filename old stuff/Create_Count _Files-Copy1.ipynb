```python
import pandas as pd
from collections import Counter
import os


def convertData_pickle(fileName, csvCat, curFileIdx):
    Words = []
    lemma = []
    pos = []
    full_file_df = pd.DataFrame([])
    with open(fileName) as f:
        f.readline()
        for line in f:
            info = line.split()
            #print(info)
            Words.append(info[0])
            lemma.append(info[1])
            pos.append(info[2])
        f.close()
    lemma_counts= dict(Counter(lemma))
    pos_counts = dict(Counter(pos))
    feature_keys = list(lemma_counts.keys())+list(pos_counts.keys())
    feature_counts = list(lemma_counts.values()) + list(pos_counts.values())
    feature_counts_df = pd.DataFrame({"features": feature_keys,"counts ": feature_counts })
    feature_counts_df = feature_counts_df.T
    ##########################################################################################    
    print(feature_counts_df.head())
    feature_counts_df.to_pickle(f"./pickle_dfs/{csvCat}_{curFileIdx}.pkl")

   

    
    

rootdir = '/Users/abigailamick/Desktop/CompSemanticsProject/text_files_centuries'
currentCentIdx= 0
curNumFile = 0
currentCent = ""
prevCent = ""
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        if "th" in subdir:
            currentCent = subdir[subdir.find("th") - 2:subdir.find("th")]
        elif "st" in subdir:
            currentCent = subdir[subdir.find("st") - 2:subdir.find("st")]
        #print(f"{currentCent}_{curNumFile}")
        convertData_pickle(os.path.join(subdir, file), currentCent, curNumFile)
        curNumFile += 1
    prevCent = currentCent





    




```
